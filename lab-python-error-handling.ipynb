# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {"t-shirt": 34, "mug": 55, "hat": 3, "book": 100, "keychain": 6}
customer_orders = set()
order_statistics = {}

def get_customer_orders():
    while True:
        order = input("Enter a product to order (t-shirt, mug, hat, book, keychain): ")
        try:
            if order not in products:
                raise ValueError("This product is not available.")
            customer_orders.add(order)         
            add = input("Do you want to add another product? (yes/no): ")
            if add == "no":
                break
            elif add!= "yes":
                raise ValueError("Invalid input. Please enter 'yes' or 'no'.")
                
        except ValueError as error:
            print(f"Error: {error}")

    print("Customer Orders:", customer_orders)
````

````python

def update_inventory (customer_orders,inventory):
    for product in customer_orders:
        inventory[product] -= 1    


def print_order_statistics (order_statistics):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    order_statistics = {"Products ordered": (total_products_ordered), "Percentage ordered": (percentage_ordered)}
    order_statistics["Products ordered"]= total_products_ordered
    order_statistics["Percentage ordered"]= percentage_ordered
    print(f"Order Statistics:")
    for k,v in order_statistics.items():
        print(k,v)


def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")


def initialize_inventory (products):
    initialize=input("Initialize inventory? yes/no")
    if initialize=="yes":
        print("Products available:")
        for product, quantity in inventory.items():
            print(product, quantity)
        get_customer_orders()
        update_inventory (customer_orders, inventory)
        print_order_statistics(order_statistics)
        print_updated_inventory(inventory)
    elif initialize!="yes" and initialize!="no":
        initialize_inventory (products)

initialize_inventory(products)
````
